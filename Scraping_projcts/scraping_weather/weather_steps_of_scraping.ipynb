```python
import requests as rq
from bs4 import BeautifulSoup as bs
import csv
```

```python
url='https://eg.freemeteo.com/weather/cairo/current-weather/location/?gid=360630&language=english&country=egypt'
response=rq.get(url,'html.parser').text
response
```

```python
soup=bs(response)
soup
```

```python
temp=soup.find('div',attrs={'class':'temp'}).get_text().split('°')[0]
temp
```

```python
wind=soup.find('span',attrs={'class':'white'}).b.get_text().strip()
wind
```

```python
details=soup.find('div',attrs={'class':'stats'}).get_text().split(' ')
details
```

```python
v_c=[]
for i in details:
    if i.endswith('%'):
        humidity=i
    if i.endswith('m'):
        v_c.append(i)
    if i.endswith('mbScattered'):
        pressure=i
    
```

```python
humidity=humidity.split('%')[0]
humidity
```

```python
visibility=v_c[0].split('m')[0]
visibility
```

```python
cloud=v_c[1].split('m')[0]
cloud
```

```python
pressure=pressure.split('m')[0]
pressure
```

```python
data={'temp':temp,'wind':wind,'humidity':humidity,'visibility':visibility,'cloud':cloud,'pressure':pressure}
data
```
