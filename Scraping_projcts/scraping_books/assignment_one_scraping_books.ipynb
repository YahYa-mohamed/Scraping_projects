```python
import csv
import requests as rq
from bs4 import BeautifulSoup as bs
import pandas as pd
```

```python
with open('books_details.csv','w') as f:
    fnames=['names','rates','prices']
    writer=csv.DictWriter(f,fieldnames=fnames)
    writer.writeheader()
    url='http://books.toscrape.com/catalogue/category/books/science_22/index.html'
    response=rq.get(url)
    response=response.text
    soup=bs(response,'html.parser')
    all_l=soup.find('ol',attrs={'class':'row'}).find_all('li')
    rate={'One':1,'Two':2,'Three':3,'Four':4,'Five':5}
    
    for i in range(0,len(all_l)):
        names=all_l[i].h3.a.get_text()
        rates=all_l[i].p.get_attribute_list('class')[1]
        prices=all_l[i].find('p',attrs={'class':'price_color'}).get_text().split('Â£')[1]
        
        data={'names':names,'rates':rate[rates],'prices':prices}
        writer.writerow(data)
       
```
